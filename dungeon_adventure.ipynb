```python
# Step 1: Player Setup
# Prompt the player for their name.
# Set health to 10.
# Create an empty list to store their inventory.
import random
import sys

Player1 = input("Enter your name: ")
Player1_health = 10
Player1_inventory = []
```

```python
# Step 2: Treasure Dictionary
# Create a dictionary of treasures where the keys are treasure names
# and the values are the gold value of each treasure.
treasure = {
    "Master Sword": 1000,
    "Heroes Shield": 750,
    "Boomerang": 50, 
    "Bombs": 10,
    "Arrows": 10,
    "Hookshot": 200
    }
```

```python
# Step 3: Game Loop - Moving through 5 rooms
# Use a for loop to represent rooms 1 to 5.
# In each room, display the current room number and the options:
# 1. Search for treasure
# 2. Move to next room
# 3. Check health and inventory
# 4. Quit the game
# Make sure to handle each option appropriately.
rooms = [1 ,2 ,3 ,4 ,5 ]

for room in rooms:
    print(f"\n You are in room {room}.")

    active_room = True
    
    while active_room:
        print("Options:")
        print("1. Search for treasure")
        print("2. Move to next room")
        print("3. Check health and inventory")
        print("4. Quit the game")

        choice = input("Choose an option (1-4): ")

# Step 4: Implement treasure search
# If the player chooses to search, randomly decide if they find treasure or a trap.
# If treasure, randomly pick from the treasure dictionary and add to inventory.
# If trap, deduct 2 health points and print a message.

    
        if choice == "1":
            outcome = random.choice(["treasure", "trap"])
            if outcome == "treasure":
                chest_treasure = random.choice(list(treasure.keys()))
                Player1_inventory.append(chest_treasure)
                print(f"You found {chest_treasure}! You saved {treasure[chest_treasure]} gold.")
            else:
                Player1_health -= 2
                print("Oh NO! Admiral Akbar was right! It was a trap!!! You lost 2 hp.")
        
        elif choice == "2":
            print("you open the next door and enter.")
            active_room = False 

        elif choice == "3":
            print(f"Health: ({Player1_health} HP)")
            print(f"Inventory: {Player1_inventory if Player1_inventory else 'Nothing'}")

        elif choice == "4":
            print("Are you sure you want to quit? Gannon's minions will join forces with the Empire and take over the galaxy!")
            confirm = input("Type 'y' or 'n'").strip().lower()
            if confirm == 'y':
                print("You have chosen to quit the game. Hyrule is doomed!")
                sys.exit()
            else:
                print("You have chosen to continue the game.")


        else:
            print("Don't be that guy and pick an option that is valid.")

# Step 5: Check health
# If health drops to 0 or below, print a Game Over message and break the loop.
        if Player1_health <= 0:
            print("Game Over! Gannon has joined the Empire! The galaxy is doomed!")
            exit()
```

```python
# Step 6: End of Game Summary
# If the player quits or finishes all rooms, calculate the total value of the collected treasures.
# Loop through the inventory and sum the gold value using the treasure dictionary.
# Print the final health, inventory, and total treasure value.
print(f"congratulations, {Player1}! You have saved Hyrule and the galaxy!!")

total_treasure_value = 0
for item_name in Player1_inventory:
    if item_name in treasure:   
        total_treasure_value += treasure[item_name]
        
print(f"Game over! {Player1} has saved Hyrule and the galaxy from the forces of Gannon and the Galactic Empire!")
print(f"Health: {Player1_health} HP")
print(f"Inventory: {Player1_inventory}")
print(f"Total value of items: {total_treasure_value}")
```
